### recomindation


```python
transport_city=data.groupby(['city','transport']).size().rename("count").reset_index().pivot(index='city',columns='transport',values='count')
plt.figure(figsize=(10,10))
transport_city.plot.barh()
```

* we recomend our custmers to use the public transportations to reach our restorant based on our data in the figure thats showng the number of peoble that use every kind of trnsportation to reach the restourant


```python
budget_city=data.groupby(['city','budget']).size().rename("count").reset_index().pivot(index='city',columns='budget',values='count')
plt.figure(figsize=(10,10))
budget_city.plot.barh()
```

```python
activity_city=data.groupby(['city','activity']).size().rename("count").reset_index().pivot(index='city',columns='activity',values='count')
activity_city.plot.barh()
```
